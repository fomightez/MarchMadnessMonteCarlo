## Here are our standard imports to make sure that we have access to the things we need

```python
from __future__ import division
import numpy as np
import pylab
import matplotlib.cm as cm
from numpy.random import random #import only one function from somewhere
from numpy.random import randint
import scipy
from time import sleep
```

## Defining a function

```python
def square(number):
    return number * number # don't forget the return!
print square(4)
```

## If statements

```python
if 2 > 4:
    print "Something is wrong in the world!"
else:
    print "Everything is happy!"
```

## Choosing a random number

```python
i = random() # random number between 0 and 1
print i
```

## Making an array of random numbers

```python
ra = random((10,10)) # array 10 by 10
```

## Plotting the array in the default style, with colors and interpolation

```python
pylab.imshow(ra)
```

## Setting ranges of values equal to something inside that array

```python
ra[ra<0.5] = -1
ra[ra>=0.5] = 1
```

## Plotting the array in black and white, with no interpolation

```python
pylab.imshow(ra,interpolation='nearest',cmap=cm.Greys_r)
```

## For loop

```python
for thing in xrange(5):
    print "thing is",thing
    print "   thing squared is",thing*thing # we include some leading spaces to make the output look nice
    print "   which is also",square(thing)
    print "   Also, I like cheese"
```

## Adding things to a list

```python
result = [] # empty list
result.append('apple')
print result
result.append('peach')
print result
# append adds one item to the end. If you want to combine lists, you use the extend function instead
result.extend(['pear','plum']) #Make sure that you pass a *list* to the extend function
print result
```
